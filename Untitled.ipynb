```python
import pygame
import sys
import random

# Инициализация Pygame
pygame.init()

# Настройки экрана
screen_width = 800
screen_height = 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Игра 'Пушка' — ловим кошек!")

# Цвета
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)
YELLOW = (255, 255, 0)

# Настройки пушки (рамки)
gun_width = 100
gun_height = 20
gun_x = screen_width // 2 - gun_width // 2
gun_y = screen_height - 50
gun_speed = 5

# Настройки кошек
cat_width = 30
cat_height = 30
cats = []
cat_speed = 3
spawn_rate = 1000  # мс между появлениями кошек
last_spawn = pygame.time.get_ticks()

# Счётчик баллов
score = 0
font = pygame.font.SysFont(None, 36)

# Часы для контроля FPS
clock = pygame.time.Clock()

# Загрузка изображения кошки (если есть файл cat.png в папке с игрой)
try:
    cat_image = pygame.image.load('cat.png')
    cat_image = pygame.transform.scale(cat_image, (cat_width, cat_height))
except:
    cat_image = None  # Если нет изображения, будем рисовать прямоугольник

# Основной игровой цикл
running = True
while running:
    # Обработка событий
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Управление пушкой (рамкой) с помощью стрелок
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and gun_x > 0:
        gun_x -= gun_speed
    if keys[pygame.K_RIGHT] and gun_x < screen_width - gun_width:
        gun_x += gun_speed

    # Создание новых кошек
    current_time = pygame.time.get_ticks()
    if current_time - last_spawn > spawn_rate:
        cat_x = random.randint(0, screen_width - cat_width)
        cat_y = 0
        cats.append(pygame.Rect(cat_x, cat_y, cat_width, cat_height))
        last_spawn = current_time

    # Движение кошек вниз
    for cat in cats:
        cat.y += cat_speed
    # Удаление кошек, улетевших за экран
    cats = [cat for cat in cats if cat.y < screen_height]

    # Проверка столкновений (поймали кошку?)
    for cat in cats:
        if cat.colliderect(pygame.Rect(gun_x, gun_y, gun_width, gun_height)):
            cats.remove(cat)
            score += 10  # Добавляем баллы за пойманную кошку

    # Очистка экрана
    screen.fill(BLACK)

    # Рисование элементов
    # Пушка (рамка)
    pygame.draw.rect(screen, BLUE, (gun_x, gun_y, gun_width, gun_height), 3)
    # Кошки
    for cat in cats:
        if cat_image:
            screen.blit(cat_image, cat)  # Рисуем изображение кошки
        else:
            pygame.draw.rect(screen, YELLOW, cat)  # Рисуем жёлтый прямоугольник
    # Счёт
    score_text = font.render(f"Счёт: {score}", True, WHITE)
    screen.blit(score_text, (10, 10))

    # Обновление экрана
    pygame.display.flip()
    clock.tick(60)  # 60 FPS

# Завершение Pygame
pygame.quit()
sys.exit()
```
